```python
import openmm
from openmm import app
from openmm.unit import *
from openmm import Vec3
```

```python
pdb = app.PDBFile('config.pdb')     # must include bonds across boundaries
modeller = app.Modeller(pdb.topology, pdb.positions)
```

```python
box = pdb.topology.getPeriodicBoxVectors()
box
```

```python
a, b, c = box
a, b, c
```

```python

orig_top = modeller.topology
orig_pos = modeller.positions
```

```python
nx, ny, nz = 8, 8, 1
```

```python
for i in range(nx):
    for j in range(ny):
        for k in range(nz):
            offset = i*a + j*b + k*c
            modeller.add(orig_top, [p + offset for p in orig_pos])
```

```python
rep_top = modeller.topology
rep_top.setPeriodicBoxVectors((nx*a.value_in_unit(nanometer), ny*b.value_in_unit(nanometer), nz*c.value_in_unit(nanometer)))
rep_pos = modeller.positions
```

```python
rep_top.getPeriodicBoxVectors()
```

```python
with open(f"silica-{nx}{ny}{nz}.pdb", "w") as f:
    app.PDBFile.writeFile(rep_top, rep_pos, f)
```
